```python
import random
import math
import numpy as np


# Инициализация области
def area_start(large,percent_red,percent_blue):
  area = np.zeros((large, large), int)
  percent_empty = 1 - percent_red - percent_blue
  area_count = large * large
  red_count = area_count* percent_red
  blue_count = area_count* percent_blue
  i = 0
  j = 0
  while i <= red_count:
    x = random.randint(0,large-1)
    y = random.randint(0,large-1)
    if(area[x][y] == 0):
      area[x][y] = int(1)
      i = i + 1
  while j <= blue_count:
    x = random.randint(0,large-1)
    y = random.randint(0,large-1)
    if(area[x][y] == 0):
      area[x][y] = int(2)
      j = j + 1
  return area

# Случайный поиск "несчастных" клеток
def find_sad(area):
  find = 0
  large = len(area)
  while(find != 1):
    neighbours_count = 0
    x = random.randint(0,large-1)
    y = random.randint(0,large-1)
    if(area[x][y]!=0):
      if(x != 0) and (y != 0) and (area[x-1][y-1] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+1")
      if(x != 0) and (area[x-1][y] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+2")
      if(x != 0) and (y != large-1) and (area[x-1][y+1] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+3")
      if(y != 0) and (area[x][y-1] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+4")
      if(y != large-1) and (area[x][y+1] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+5")
      if(x != large-1) and (y != 0) and (area[x+1][y-1] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+6")
      if(x != large-1) and (area[x+1][y] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+7")
      if(x != large-1) and (y != large-1) and (area[x+1][y+1] == area[x][y]):
        neighbours_count = neighbours_count + 1
        #print("+8")

      if(neighbours_count < 2):
        find = 1
        print("'Грустная' клетка: ")
        print("X: ", x,"Y: ", y,"Значение: ", area[x][y], "Количество сосед. клеток того же цвета: ", neighbours_count)

  return x,y

# Случайный поиск пустых клеток
def find_empty(area):
  find = 0
  large = len(area)
  while(find != 1):
    x = random.randint(0,large-1)
    y = random.randint(0,large-1)
    if(area[x][y]==0):
      find = 1
      print("Пустая клетка: ")
      print("X: ", x,"Y: ", y,"Значение: ", area[x][y])
  return x,y

# Перемещение клеток
def replace(area,x1,y1,x2,y2):
  area[x2][y2] = area[x1][y1]
  area[x1][y1] = 0



# Основное тело программы
large = int(input())
percent_red = 0.45
percent_blue = 0.45
area = area_start(large,percent_red,percent_blue)
print("Начальная область: \n", area)

for i in range (200):
  print("Номер итерации: ", i+1)
  x1,y1 = find_sad(area)
  x2,y2 = find_empty(area)
  replace(area,x1,y1,x2,y2)
  if(i % 50 == 0):
    print(area)
print("Итог. область: \n", area)


```
